```python
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("DF FOR HIVE")\
                            .config("spark.sql.warehouse.dir", "hdfs://localhost:54310/user/hive/warehouse")\
                            .enableHiveSupport()\
                            .getOrCreate()
```
